# Using centroid as obstacle distance metric:

```python
q0 = (0,0)
qgoal = (10,10)
Q_star = 20
eta = 4
d_star = 5
zeta = 0.27
obs1 = Obstacle((1,1),1,4,c=False,circ=True, Q_star=Q_star, eta=eta*1.5)
obs2 = Obstacle((3,4),1,8,c=False,circ=True, Q_star=Q_star, eta=eta)
obs3 = Obstacle((3,12),9,1,c=False,circ=True, Q_star=Q_star, eta=eta*2)
obs4 = Obstacle((12,5),1,8,c=False,circ=True, Q_star=Q_star, eta=eta*0.1)
obs5 = Obstacle((6,5),6,1,c=False,circ=True, Q_star=Q_star*0.5, eta=eta*5)
obstacles = [obs1,obs2,obs3,obs4,obs5]
RGD = RectGradDescent(q0, qgoal, obstacles, max_iter=50,d_star = d_star, zeta=zeta)
```

```python
RGD.plot_v_field()
```

```python
RGD.plot_path()
print("Path Length:",RGD.path_length)
```
