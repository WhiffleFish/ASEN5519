```python
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
import pandas as pd
```

```python
s_results = pickle.load(open("PickleData/s_results1.p",'rb'))
ns_results = pickle.load(open("PickleData/ns_results1.p",'rb'))

benchmark_vars = [(200, 0.5), (200, 1), (200, 1.5), (200, 2), (500, 0.5), (500, 1), (500, 1.5), (500, 2)]
distances = []
comptimes = []
smootheds = []
benchmarknums = []
bm_ind = []
for i, test in enumerate(ns_results):
    benchmarknums.extend([benchmark_vars[i]]*100)
    bm_ind.extend([i]*100)
    smootheds.extend([False]*100)
    distances.extend(ns_results[i]['distance'])
    comptimes.extend(ns_results[i]['time_elapsed'])
    
for i, test in enumerate(s_results):
    benchmarknums.extend([benchmark_vars[i]]*100)
    bm_ind.extend([i]*100)
    smootheds.extend([True]*100)
    distances.extend(s_results[i]['distance'])
    comptimes.extend(s_results[i]['time_elapsed'])
    
data_dict = {'dist':distances,'time':comptimes,'smooth':smootheds,'benchmark':benchmarknums, 'bm_ind':bm_ind}
df = pd.DataFrame(data_dict)

fig = plt.figure(constrained_layout=True, figsize=(15,15))
gs = fig.add_gridspec(4,4)
ax1 = fig.add_subplot(gs[0:3,0:2])
sns.boxplot(x='benchmark',y='time',data=df, hue='smooth',ax=ax1)
ax1.set_ylabel("Computation Time (s)", fontsize=12)

ax2 = fig.add_subplot(gs[0:3,2:4])
sns.boxplot(x='benchmark',y='dist',data=df, hue='smooth',ax=ax2)
ax2.set_ylabel("Path Distance", fontsize=12)

ax3 = fig.add_subplot(gs[3:4,1:3])
ax3.bar(x=list(range(len(benchmark_vars))),height=[df[df['bm_ind'] == i]['dist'].isnull().sum()/2 for i in range(len(benchmark_vars))])
ax3.set_xlabel("benchmark #")
ax3.set_ylabel("Percentage of Pathfinding Failures", fontsize=12)
ax3.grid(True,axis='y')
fig.suptitle("Workspace 1 Benchmark", fontsize=20)
```
